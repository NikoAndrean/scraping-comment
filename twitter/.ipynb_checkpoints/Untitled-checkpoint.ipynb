```python
# df = pd.DataFrame(twitter_data)
# df.to_json('tweets.json', orient ='index')
# print('file exported')
while tweets_retrieved < total_tweets_needed:
    # Set parameter offset untuk setiap halaman
    payload['offset'] = offset
    response = requests.get(
        'https://api.scraperapi.com/structured/twitter/search', params=payload)
    
    # Periksa apakah respons berhasil
    if response.status_code != 200:
        print("Error:", response.status_code)
        break
    
    data = response.json()
    
    # Ambil hasil organik (tweets) dari respons
    all_tweets = data.get('organic_results', [])
    if not all_tweets:
        break
    
    # Tambahkan tweet ke list dan perbarui counter
    twitter_data.extend(all_tweets)
    tweets_retrieved += len(all_tweets)
    offset += len(all_tweets)

# Buat DataFrame dan ekspor ke JSON
df = pd.DataFrame(twitter_data[:total_tweets_needed])  # Mengambil hanya 1000 data
df.to_json('tweets.json', orient='index')
print("File 'tweets.json' telah diekspor dengan jumlah data:", len(df))
```
